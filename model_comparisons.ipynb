### Nested cross validation for elasticNet model

```python
# Define the pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', ElasticNet(max_iter=10000))
])

# Hyperparameter grid
param_grid = {
    'model__alpha': [0.01, 0.1, 1.0],
    'model__l1_ratio': [0.1, 0.5, 0.9]
}

# Inner CV: tuning
inner_cv = KFold(n_splits=5, shuffle=True, random_state=1)
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=inner_cv,
    scoring=make_scorer(mean_absolute_error, greater_is_better=False),
    n_jobs=-1
)

# Outer CV: evaluation
outer_cv = KFold(n_splits=5, shuffle=True, random_state=2)
nested_scores = cross_val_score(
    grid_search,
    train_sample, train_age_sample,
    cv=outer_cv,
    scoring=make_scorer(mean_absolute_error, greater_is_better=False),
    n_jobs=-1
)

print(f"Nested CV MAE: {-nested_scores.mean():.3f} ± {nested_scores.std():.3f}")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from scikeras.wrappers import KerasRegressor
from sklearn.base import BaseEstimator, RegressorMixin
from sklearn.model_selection import KFold, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import make_scorer, mean_absolute_error

# === 1. Define the Keras model ===
def build_model(learning_rate=0.001, dropout_rate=0.0, **kwargs):
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1))  # Regression output
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss='mean_absolute_error')
    return model

# === 3. Create the pipeline ===
regressor = KerasRegressor(
    model=build_model,
    learning_rate=0.001,
    dropout_rate=0.0,
    epochs=50,
    batch_size=32,
    verbose=0
)

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('regressor', regressor)
])

# === 4. Define inner CV parameter grid ===
param_grid = {
    'regressor__learning_rate': [0.001, 0.01],
    'regressor__dropout_rate': [0.0, 0.3]
}

# === 5. Set up the inner CV ===
inner_cv = KFold(n_splits=3, shuffle=True, random_state=1)

# === 6. Set up the outer CV ===
outer_cv = KFold(n_splits=5, shuffle=True, random_state=42)

# === 7. GridSearchCV inside outer cross-validation ===
grid = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=inner_cv,
    scoring=make_scorer(mean_absolute_error, greater_is_better=False),
    n_jobs=-1  # Parallelization (set to -1 for maximum)
)

# === 8. Perform Nested Cross-Validation ===
nested_scores = cross_val_score(
    grid,
    X_sample, y_sample,
    cv=outer_cv,
    scoring=make_scorer(mean_absolute_error, greater_is_better=False),
    n_jobs=-1
)

# Output the mean and standard deviation of MAE from the nested cross-validation
print(f"Nested CV MAE: {-nested_scores.mean():.3f} ± {nested_scores.std():.3f}")
```
